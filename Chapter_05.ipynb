## 7.จงอธิบายความแตกต่างระหว่างลิสต์และทัพเพิล

```python
"""
คำตอบ
"""
-------------------------------------------------------------------------------
List (ลิสต์)
การประกาศ - ใช้เครื่องหมายวงเล็บเหลี่ยม [ ]
การเปลี่ยนแปลงค่า (Mutable) - เปลี่ยนแปลงได้ เช่น เพิ่ม ลบ หรือแก้ไขสมาชิก
การใช้งานทั่วไป - ใช้เก็บข้อมูลที่ต้องการเปลี่ยนแปลงบ่อย
เช่น my_list = [10, 20, 30]
การเข้าถึงสมาชิก = my_list[0]
-------------------------------------------------------------------------------
Tuple (ทัพเพิล)
การประกาศ - ใช้เครื่องหมายวงเล็บปกติ ( )
การเปลี่ยนแปลงค่า (Mutable) - เปลี่ยนแปลงไม่ได้ (Immutable) หลังประกาศแล้วจะแก้ไขค่าไม่ได้
การใช้งานทั่วไป - ใช้เก็บข้อมูลที่ไม่ต้องการเปลี่ยนแปลง เช่น ค่าคงที่
เช่น my_tuple = (10, 20, 30)
ใช้ดัชนีเช่นเดียวกัน = my_tuple[0]
-------------------------------------------------------------------------------
```

## 8.สลับค่าตัวแปร

```python
a = 10
b = 20

print ("a : {} , b: {}".format(a, b))

#ผลที่คาดหวังคือ|
# a:20 , b: 10
```

```python
#คำตอบ
a = 10
b = 20

tmp = a
a = b
b = tmp

print ("a : {} , b: {}".format(a, b))
```

```python
a = 10
b = 20
#parallel assignment
a, b = b, a

print ("a : {} , b: {}".format(a, b))
```
